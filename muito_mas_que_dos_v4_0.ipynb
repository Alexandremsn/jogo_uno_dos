```python
import random
"""importa biblioteca usada no jogo"""
print("######  #######  ##### " )
print('#     # #     # #     # ')
print('#     # #     # #       ')
print('#     # #     #  #####  ')
print('#     # #     #       # ')
print('#     # #     # #     # ')
print('######  #######  #####  ')
print('Vamos jogar DOS um jogo muito melhor que UNO')
"""imprime o título do jogo"""

#formatar texto das regras
deck_principal = [item for item in range(30)]
'''gera o baralho de cartas'''
dict_cartas = {0: ['amarelo', 'proibido'], 1: ['amarelo', '1'], 2: ['amarelo', '2'], 3: ['amarelo', '3'],
               4: ['amarelo', '4'], 5: ['amarelo', '5'],
               6: ['amarelo', '6'], 7: ['amarelo', '7'], 8: ['amarelo', '8'], 9: ['amarelo', 'Mas DOS'],
               10: ['vermelho', 'proibido'], 11: ['vermelho', '1'], 12: ['vermelho', '2'], 13: ['vermelho', '3'],
               14: ['vermelho', '4'], 15: ['vermelho', '5'],
               16: ['vermelho', '6'], 17: ['vermelho', '7'], 18: ['vermelho', '8'], 19: ['vermelho', 'Mas DOS'],
               20: ['laranja', 'proibido'], 21: ['laranja', '1'], 22: ['laranja', '2'], 23: ['laranja', '3'],
               24: ['laranja', '4'], 25: ['laranja', '5'],
               26: ['laranja', '6'], 27: ['laranja', '7'], 28: ['laranja', '8'], 29: ['laranja', 'Mas DOS']}
'''cria um dicionário de equivalências de numero da carta e seus atributos'''

print('O jogo possui as cartas com 3 cores, amarelo,  vermelho  e  laranja. Os números vão de 1 a 8.') 
print('A carta  Proibido, faz o oponente perder a vez e a carta  Mas Dos faz o oponente comprar duas') 
print('cartas e perder a vez.   Cada jogador recebe 5 cartas.   É sorteada uma carta para a mesa que')
print('não apresenta efeito,  serve apenas para determinar a  cor e  número iniciais. O jogador deve')
print('colocar uma carta na mesa da mesma  cor ou com o mesmo número/tipo,  esta carta  se tornará a')
print('nova carta da mesa.  O proximo jogador deve obedecer as especificações desta carta vence quem')
print('ficar com duas cartas ou se acabarem as cartas do baralho quem tiver menos cartas.')
'''imprime as regras do jogo'''

variavel_sem_sentido = input('Para Começar digite o seu Nome:')
print('Humm... \nEste nome rima com derrota!!!!')
print('Eu serei seu adversário\nVamos começar...')
'''como uma forma de piada ele pergunta o nome do jogador e avisa o início do jogo'''

def lcd(deck):
  """esta função le uma lista de cartas e as decodifica em um ASCII art"""
  dict_painel = {'1':['#    #     # ' , '#   ##     # ' , '#  # #     # ' , '#    #     # ' , '#    #     # ' , '#    #     # ' , '#  #####   # '],
                 '2':['#   #####  # ' , '#  #     # # ' , '#        # # ' , '#   #####  # ' , '#  #       # ' , '#  #       # ' , '#  ####### # '],
                 '3':['#   #####  # ' , '#  #     # # ' , '#        # # ' , '#   #####  # ' , '#        # # ' , '#  #     # # ' , '#   #####  # '],
                 '4':['#  #       # ' , '#  #    #  # ' , '#  #    #  # ' , '#  #    #  # ' , '#  ####### # ' , '#       #  # ' , '#       #  # '],
                 '5':['#  ####### # ' , '#  #       # ' , '#  #       # ' , '#  ######  # ' , '#        # # ' , '#  #     # # ' , '#   #####  # '],
                 '6':['#   #####  # ' , '#  #     # # ' , '#  #       # ' , '#  ######  # ' , '#  #     # # ' , '#  #     # # ' , '#   #####  # '],
                 '7':['#  ####### # ' , '#  #    #  # ' , '#      #   # ' , '#    #     # ' , '#   #      # ' , '#   #      # ' , '#   #      # '],
                 '8':['#   #####  # ' , '#  #     # # ' , '#  #     # # ' , '#   #####  # ' , '#  #     # # ' , '#  #     # # ' , '#   #####  # '],
                 'proibido':['#    ###   # ' , '#   #   #  # ' , '#  # #   # # ' , '#  #  #  # # ' , '#  #   # # # ' , '#   #   #  # ' , '#    ###   # '],
                 'Mas DOS':['#   #####  # ' , '#  # MAS # # ' , '#    DOS # # ' , '#   #####  # ' , '#  # MAS   # ' , '#  # DOS   # ' , '#  ####### # '],
                 'texto':['############ ' , '# VERMELHO # ', '# AMARELO  # ', '# LARANJA  # ','#          # ','# posicao','# ']}
  linha_00 = ''
  linha_01 = ''
  linha_02 = ''
  linha_03 = ''
  linha_04 = ''
  linha_05 = ''
  linha_06 = ''
  linha_07 = ''
  linha_08 = ''
  linha_09 = ''
  linha_10 = ''
  cont = 0
  '''acima criamos um dict com os paineis de cada carta e as varívaveis que guardam as informações de cada linha da  ASCII art'''
  for carta in deck:
    numero_da_carta = dict_cartas[carta][1]
    collor = dict_cartas[carta][0]
    linha_00 += dict_painel['texto'][0]
    linha_10 += dict_painel['texto'][0]
    cont += 1
    if cont >= 10:
      linha_posicao_espaco = ''
    else:
      linha_posicao_espaco = ' '
    linha_posicao_contador = dict_painel['texto'][5]+linha_posicao_espaco+str(cont)+dict_painel['texto'][6]
    if(deck == pote):
      linha_posicao_contador = dict_painel['texto'][4]
    linha_09 += linha_posicao_contador
    if collor == 'vermelho':
        linha_08+= dict_painel['texto'][1]
    elif collor == 'amarelo':
        linha_08+= dict_painel['texto'][2]
    elif collor == 'laranja':
        linha_08+= dict_painel['texto'][3]
    linha_01 += dict_painel[numero_da_carta][0]
    linha_02 += dict_painel[numero_da_carta][1]
    linha_03 += dict_painel[numero_da_carta][2]
    linha_04 += dict_painel[numero_da_carta][3]
    linha_05 += dict_painel[numero_da_carta][4]
    linha_06 += dict_painel[numero_da_carta][5]
    linha_07 += dict_painel[numero_da_carta][6]
  '''convertemos as cartas em linha de string e adicionarmos a linha do painel buscando suas equivalências nos dois dicionários'''
  print(linha_00)
  print(linha_01)
  print(linha_02)
  print(linha_03)
  print(linha_04)
  print(linha_05)
  print(linha_06)
  print(linha_07)
  print(linha_08)
  print(linha_09)
  print(linha_10)


def apresenta(deck):
    '''esta função converte os números das caras para sua cor e número'''
    if type(deck) == list:
        deck = deck[0]
    
    lista_apresenta = dict_cartas[deck][1] + ' ' + dict_cartas[deck][0]
    '''na parte final ele monta um string para ser usado quando exibir cartas jogadas'''
    '''anexando isso em string e retornando'''
    return lista_apresenta


print('Dando as cartas....')
cartas_dadas = random.sample(range(0, 29), 10)
deck_computador = [cartas_dadas[0], cartas_dadas[1], cartas_dadas[2], cartas_dadas[3], cartas_dadas[4]]
deck_jogador = [cartas_dadas[5], cartas_dadas[6], cartas_dadas[7], cartas_dadas[8], cartas_dadas[9]]
'''ele dá as cartas, divide elas e depois as apaga da lista do baralho'''

for i in deck_computador:
    deck_principal.remove(i)
for i in deck_jogador:
    deck_principal.remove(i)

pote = random.sample(deck_principal, 1)

deck_principal.remove(pote[0])


def jogador_joga():
    print(f'O Computador tem {len(deck_computador)} cartas')
    print('Sua Mão:')
    lcd(deck_jogador)
    print('A Carta da mesa é:')
    lcd(pote)
    '''a principal função do jogo exibe as cartas do jogador a carta da mesa e quantas cartas o computador tem na mão'''
    print('digite a posição da carta na sua mão ou digite 0 para comprar:')
    possivel = [str(item) for item in range(29)]
    validador_jogada = 0
    jogada = input('digite:')
    '''abaixo validamos a jogada informada está válida e dentro do range'''

    while validador_jogada == 0:
        if jogada in possivel:
            jogada = int(jogada)-1
            if jogada+1 > len(deck_jogador):
                print('Jogada Inválida')
                jogada = input('digite:')
            else:
                validador_jogada = 1
        else:
            print('Entrada Inválida!!!')
            jogada = input('digite:')

    padrao = dict_cartas[pote[0]]
    selecionada = dict_cartas[deck_jogador[jogada]]
    '''uma vez validada a jogada é computada e convertida de número da carta para informação da carta'''
    print('Carta Selecionada:', apresenta(deck_jogador[jogada]))
    cartas_compradas = []
    if jogada != -1 and ((selecionada[0] == padrao[0]) or (selecionada[1] == padrao[1])):
        if selecionada[1] == 'proibido':
            '''se a carta for do tipo proibido ela deve pular a vez do oponente assim invocamos mais uma vez a função de jogar após computar a jogada'''
            del (pote[0])
            print('computador perdeu a vez')
            pote.append(deck_jogador[jogada])
            deck_jogador.remove(deck_jogador[jogada])
            if len(deck_jogador) <= 2:
                return 'end'
            jogador_joga()
        elif selecionada[1] == 'Mas DOS':
            del (pote[0])
            print('Computador Compra duas cartas e perde a vez')
            '''esta função executa a mas dos ela deve além de pular a vez sortear mais duas cartas do baralho e adicionar ao baralho do oponente'''
            pote.append(deck_jogador[jogada])
            deck_jogador.remove(deck_jogador[jogada])
            cartas_compradas = random.sample(deck_principal, 2)
            deck_computador.append(cartas_compradas[0])
            deck_computador.append(cartas_compradas[1])
            deck_principal.remove(cartas_compradas[0])
            deck_principal.remove(cartas_compradas[1])
            print(f'computador agora tem {len(deck_computador)} cartas')
            cartas_compradas = []
            if len(deck_jogador) <= 2:
                return 'end'
            jogador_joga()
        else:
            del (pote[0])
            pote.append(deck_jogador[jogada])
            deck_jogador.remove(deck_jogador[jogada])
    elif jogada == -1:
        comprada = random.sample(deck_principal, 1)
        deck_jogador.append(comprada[0])
        deck_principal.remove(comprada[0])
        del (comprada[0])
        '''esta parte do código é ativada quando o jogador coloca zero como entrada asim sorteia uma carta e adicona ao baralho do jogador'''

    else:
        print('Jogada Inválida')
        jogador_joga()


def computador_joga():
    '''esta função executa a jogada automática do oponente'''
    padrao = dict_cartas[pote[0]]
    marcador = 0
    index = 'a'
    for i in deck_computador:
        selecionada = dict_cartas[i]
        if (selecionada[0] == padrao[0]) or (selecionada[1] == padrao[1]):
            index = i
            marcador = 1
            print('compudador joga:')
            print(apresenta(index))

            break
        '''este laço testa se o computador tem alguma carta possível para jogar se ele encontrar ele seleciona a carta e interrompe o laço'''
    if selecionada[1] == 'proibido' and marcador == 1:
        del (pote[0])
        pote.append(index)
        deck_computador.remove(index)
        print('computador jogou:', apresenta(pote), 'Você perdeu a vez!!!')
        marcador = 0
        if len(deck_computador) <= 2:
            return 'end'
        computador_joga()
        '''este trecho aplica o efeito se a carta selecionada for proibido'''
    elif selecionada[1] == 'Mas DOS' and marcador == 1:
        del (pote[0])
        pote.append(index)
        deck_computador.remove(index)
        print('computador jogou:', apresenta(pote)[0][4:], 'Você comprou duas cartas e  perdeu a vez!!!')
        cartas_compradas = random.sample(deck_principal, 2)
        deck_jogador.append(cartas_compradas[0])
        deck_jogador.append(cartas_compradas[1])
        deck_principal.remove(cartas_compradas[0])
        deck_principal.remove(cartas_compradas[1])
        del (cartas_compradas[0:1])
        marcador = 0
        if len(deck_computador) <= 2:
            return 'end'
        computador_joga()
        '''este trecho aplica o efeito da mas dos'''
    elif marcador == 1:
        del (pote[0])
        pote.append(index)
        deck_computador.remove(index)
        '''este trecho vefica se a carta selecionada foi comum e aplica o efeito da jogada comum'''
    else:
        comprada = random.sample(deck_principal, 1)
        deck_computador.append(comprada[0])
        deck_principal.remove(comprada[0])
        del (comprada[0])
        print('computador comprou uma carta')
        '''se não for jogada uma carta então o este trecho compra uma carta'''


def partida():
    '''o laço inicia e termina com o término do baralho ou algum jogador chegar a duas cartas'''
    while deck_principal:
        jogador_joga()
        if len(deck_jogador) <= 2:
            break
        elif len(deck_computador) <= 2:
            break
        '''executa a função de jodaga do jogador e testa se ele venceu ao término da funcçao'''
        computador_joga()
        if len(deck_jogador) <= 2:
            break
        elif len(deck_computador) <= 2:
            break
        '''executa a função de jogada do computador e testa se o computador venceu'''
    if len(deck_jogador) < len(deck_computador):
        print('Parabéns Você Venceu!!!!')
    elif len(deck_jogador) > len(deck_computador):
        print('Você Perdeu!!!!')
    else:
        print('Temos um Empate!!!!!')
    '''finalizado o laço ele verifica quem tem menos cartas e declara o vencedor e em caso de ingualdade declara empate'''


partida()
```
